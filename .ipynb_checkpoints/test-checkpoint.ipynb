```python
import glob
import math
import cv2
import numpy as np
import os
import matplotlib.pyplot as plt
from sklearn import preprocessing
import zipfile
```

```python
faces = {}
with zipfile.ZipFile("archive.zip") as facezip:
    for filename in facezip.namelist():
        if not filename.endswith(".pgm"):
            continue # not a face picture
        with facezip.open(filename) as image:
            # If we extracted files from zip, we can use cv2.imread(filename) instead
            faces[filename] = cv2.imdecode(np.frombuffer(image.read(), np.uint8), cv2.IMREAD_GRAYSCALE)
```

```python

fig, axes = plt.subplots(4,4,sharex=True,sharey=True,figsize=(8,10))
faceimages = list(faces.values())[-16:] # take last 16 images
for i in range(16):
    axes[i%4][i//4].imshow(faceimages[i], cmap="gray")
plt.show()
```

```python
faceshape = list(faces.values())[0].shape
print("Face image shape:", faceshape)
```

```python
print(list(faces.keys())[:5])
```

```python
classes = set(filename.split("/")[0] for filename in faces.keys())
print("Number of classes:", len(classes))
print("Number of pictures:", len(faces))
```

```python
# Take classes 1-39 for eigenfaces, keep entire class 40 and
# image 10 of class 39 as out-of-sample test
facematrix = []
facelabel = []
for key,val in faces.items():
    if key.startswith("s40/"):
        continue # this is our test set
    if key == "s39/10.pgm":
        continue # this is our test set
    facematrix.append(val.flatten())
    facelabel.append(key.split("/")[0])
 
# Create facematrix as (n_samples,n_pixels) matrix
facematrix = np.array(facematrix)
print(facematrix.shape)
```

```python
A = np.resize(facematrix, (389, 112*92)) #convert the images into vectors. Each row has an image vector. i.e. samples x image_vector matrix
mean_vector = np.sum(A, axis=0, dtype='float64')/389 #calculate the mean vector
mean_matrix = np.tile(mean_vector, (389, 1)) #make a 400 copy of the same vector. 400 x image_vector_size matrix.
A_tilde = A - mean_matrix #mean-subtracted image vectors
plt.imshow(np.resize(mean_vector, (112,92)),cmap="gray") #show the mean image vector
plt.show()
```

```python
print(A_tilde)
```

```python
L = (A_tilde.dot(A_tilde.T))/389 #since each row is an image vector (unlike in the notes, L = (A_tilde)(A_tilde.T) instead of L = (A_tilde.T)(A_tilde)
print("L shape : ", L.shape)
eigenvalues, eigenvectors = np.linalg.eig(L) #find the eigenvalues and the eigenvectors of L
idx = eigenvalues.argsort()[::-1] #get the indices of the eigenvalues by its value. Descending order.
eigenvalues = eigenvalues[idx] 
eigenvectors = eigenvectors[:, idx] #sorted eigenvalues and eigenvectors in descending order
```

```python
print(L)
```

```python
eigenvectors_C = A_tilde.T @ eigenvectors #linear combination of each column of A_tilde
eigenvectors_C.shape #each column is an eigenvector of C where C = (A_tilde.T)(A_tilde). NOTE : in the notes, C = (A_tilde)(A_tilde.T)
```

```python
eigenfaces = preprocessing.normalize(eigenvectors_C.T) #normalize only accepts matrix with n_samples, n_feature. Hence the transpose.
eigenfaces.shape
```

```python
query = faces["s40/1.pgm"].reshape(1,-1)
# test_img = cv2.resize(test_img, (112,92)) #resize the testing image. cv2 resize by width and height.
# mean_subracted_testimg = np.reshape(test_img, (test_img.shape[0]*test_img.shape[1])) - mean_vector #subtract the mean
unknown_face_vector = np.array(query, dtype='float64').flatten()
normalised_uface_vector = np.subtract(unknown_face_vector,mean_vector)
plt.imshow(np.reshape(normalised_uface_vector, (112,92)), cmap='gray')
plt.title("Mean Subtracted Test Image")
plt.show()
```

```python
q = 350 #number of chosen eigenfaces
omega = eigenfaces[:q].dot(normalised_uface_vector) #the vector that represents the image with respect to the eigenfaces.
omega.shape
```

```python
# normalised_uface_vector = np.subtract(unknown_face_vector,mean_face)
# w_unknown = np.dot(proj_data, unknown_face_vector)
```

```python
reconstructed = eigenfaces[:q].T.dot(omega) #image reconstructed using q eigenfaces.
reconstructed.shape
```

```python
alpha_1 = 3000 #chosen threshold for face detection

projected_new_img_vector = eigenfaces[:q].T @ omega #n^2 vector of the new face image represented as the linear combination of the chosen eigenfaces
diff = normalised_uface_vector - projected_new_img_vector 
beta = math.sqrt(diff.dot(diff)) #distance between the original face image vector and the projected vector.

if beta < alpha_1:
    print("Face detected in the image! ", beta)
else:
    print("No face detected in the image! ", beta)
```

```python
alpha_2 = 3000 #chosen threshold for face recognition
smallest_value = None #to keep track of the smallest value
index = None #to keep track of the class that produces the smallest value
for k in range(389):
    omega_k = eigenfaces[:q].dot(A_tilde[k]) #calculate the vectors of the images in the dataset and represent 
    diff = omega - omega_k
    epsilon_k = math.sqrt(diff.dot(diff))
    if smallest_value == None:
        smallest_value = epsilon_k
        index = k
    if smallest_value > epsilon_k:
        smallest_value = epsilon_k
        index = k
        
if smallest_value < alpha_2:
    fig, axes = plt.subplots(1,2,sharex=True,sharey=True,figsize=(8,6))
    axes[0].imshow(unknown_face_vector.reshape(112,92), cmap="gray")
    axes[0].set_title("Query")
    axes[1].imshow(facematrix[index].reshape(faceshape), cmap="gray")
    axes[1].set_title("Best match")
    plt.show()
else:
    print(smallest_value, "Unknown Face!")
```

```python
##Get Mean Face##
mean_face = np.zeros((1,112*92))

for i in facematrix:
    mean_face = np.add(mean_face,i)

mean_face = np.divide(mean_face,389).flatten()
```

```python
eig_pairs = [(eigenvalues[index], eigenvectors[:,index]) for index in range(len(eigenvalues))]
# Sort the eigen pairs in descending order:
eig_pairs.sort(reverse=True)
eigvalues_sort  = [eig_pairs[index][0] for index in range(len(eigenvalues))]
eigvectors_sort = [eig_pairs[index][1] for index in range(len(eigenvalues))]
```

```python
var_comp_sum = np.cumsum(eigvalues_sort)/sum(eigvalues_sort)
num_comp = range(1,len(eigvalues_sort)+1)
```

```python
# Test on out-of-sample image of existing class
query = faces["s39/10.pgm"].reshape(1,-1)
unknown_face_vector = np.array(query, dtype='float64').flatten()
normalised_uface_vector = np.subtract(unknown_face_vector,mean_face)
w_unknown = np.dot(proj_data, normalised_uface_vector)
euclidean_distance = np.linalg.norm(w - w_unknown, axis=1)
best_match = np.argmin(euclidean_distance)
# Visualize
fig, axes = plt.subplots(1,2,sharex=True,sharey=True,figsize=(8,6))
axes[0].imshow(unknown_face_vector.reshape(112,92), cmap="gray")
axes[0].set_title("Query")
axes[1].imshow(facematrix[best_match].reshape(faceshape), cmap="gray")
axes[1].set_title("Best match")
plt.show()
```

```python
# Apply PCA to extract eigenfaces
from sklearn.decomposition import PCA
 
pca = PCA().fit(facematrix)
```

```python
print(pca.explained_variance_ratio_)
```

```python
# Take the first K principal components as eigenfaces
n_components = 50
eigenfaces = pca.components_[:n_components]
 
# Show the first 16 eigenfaces
fig, axes = plt.subplots(4,4,sharex=True,sharey=True,figsize=(8,10))
for i in range(16):
    axes[i%4][i//4].imshow(eigenfaces[i].reshape(faceshape), cmap="gray")
plt.show()
```

```python
# Generate weights as a KxN matrix where K is the number of eigenfaces and N the number of samples
weights = eigenfaces @ (facematrix - pca.mean_).T
```

```python
weights = []
for i in range(facematrix.shape[0]):
    weight = []
    for j in range(n_components):
        w = eigenfaces[j] @ (facematrix[i] - pca.mean_)
        weight.append(w)
    weights.append(weight)
    
```

```python
weights
```

```python

# Test on out-of-sample image of existing class
query = faces["s39/10.pgm"].reshape(1,-1)
query_weight = eigenfaces @ (query - pca.mean_).T
euclidean_distance = np.linalg.norm(weights - query_weight, axis=0)
best_match = np.argmin(euclidean_distance)
print("Best match %s with Euclidean distance %f" % (facelabel[best_match], euclidean_distance[best_match]))
# Visualize
fig, axes = plt.subplots(1,2,sharex=True,sharey=True,figsize=(8,6))
axes[0].imshow(query.reshape(faceshape), cmap="gray")
axes[0].set_title("Query")
axes[1].imshow(facematrix[best_match].reshape(faceshape), cmap="gray")
axes[1].set_title("Best match")
plt.show()
```

```python
# Test on out-of-sample image of new class
query = faces["s40/1.pgm"].reshape(1,-1)
query_weight = eigenfaces @ (query - pca.mean_).T
euclidean_distance = np.linalg.norm(weights - query_weight, axis=0)
best_match = np.argmin(euclidean_distance)
print("Best match %s with Euclidean distance %f" % (facelabel[best_match], euclidean_distance[best_match]))
# Visualize
fig, axes = plt.subplots(1,2,sharex=True,sharey=True,figsize=(8,6))
axes[0].imshow(query.reshape(faceshape), cmap="gray")
axes[0].set_title("Query")
axes[1].imshow(facematrix[best_match].reshape(faceshape), cmap="gray")
axes[1].set_title("Best match")
plt.show()
```
